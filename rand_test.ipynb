````python
import numpy as np
import collections
import matplotlib.pyplot as plt
import collections
from scipy.stats import chi2

def uniform_distribution_test(random_array):
    bins = np.arange(-0.5, 10.5, 1)  # ビンの境界を設定
    plt.hist(random_array, bins=bins, density=True, rwidth=0.8)  # rwidthで棒の幅を調整
    plt.title('Uniform Distribution Test')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.xticks(np.arange(0, 10))  # x軸の目盛りを整数値に設定
    plt.show()

def poker_test_with_chi_square_test(random_array, k=5):
    counts = collections.defaultdict(int)
    for i in range(len(random_array) - k + 1):
        subseq = ''.join(map(str, random_array[i:i+k]))
        counts[subseq] += 1
    
    num_subseq = len(random_array) - k + 1
    expected_count = num_subseq / (10 ** k)
    chi_squared = sum(((count - expected_count) ** 2) / expected_count for count in counts.values())
    
    # Calculate degrees of freedom
    degrees_of_freedom = 10 ** k - 1
    
    # Calculate p-value
    p_value = chi2.cdf(chi_squared, degrees_of_freedom)
    
    print(f'Poker Test Chi-Squared: {chi_squared}')
    print(f'Degrees of freedom: {degrees_of_freedom}')
    print(f'p-value: {p_value}')

def consecutive_runs_test(random_array):
    up_runs, down_runs = 0, 0
    for i in range(1, len(random_array)):
        if random_array[i] > random_array[i-1]:
            up_runs += 1
        elif random_array[i] < random_array[i-1]:
            down_runs += 1
    
    print(f'Consecutive Runs Test: Up Runs: {up_runs}, Down Runs: {down_runs}')
````

````python
# 乱数配列の例（ここに疑似乱数配列を入れてください）
random_array = np.random.randint(0, 10, size=1000)

uniform_distribution_test(random_array)
poker_test_with_chi_square_test(random_array)
consecutive_runs_test(random_array)
````

````python
import threading

array_length = 1000

def gpt_random_array():
    import openai
    import os
    import re
    openai.api_key = "***"

    sequence = f"x_1, x_2, ..., x_[{array_length}]"
    prompt = f"""Without using any programming or algorithm, try to output the following sequence using your own thinking:
Separate the numbers with ", " and do not include line breaks.

```
Let {sequence} be a sequence of random integers, where each x_i is an integer between 0 and 9 inclusive.
```
"""


    CHAT_API_PARAMS = {
        "temperature": 0.9,
        "max_tokens": array_length*2,
        "model": "gpt-3.5-turbo"
    }
    response = openai.ChatCompletion.create(
                        messages=[{"role" : "user", "content" : prompt}],
                        **CHAT_API_PARAMS
                    )

    answer = response["choices"][0]["message"]["content"]
    print(f"GPT response: {answer}")

    array = re.findall(r'\d+', answer)
    return array


threads = []
results = []

for i in range(3):
    t = threading.Thread(target=lambda r: r.append(gpt_random_array()), args=(results,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

random_array = []
for result in results:
    random_array.extend(result)

print(f'array: {random_array}\nlength: {len(random_array)}')
    
uniform_distribution_test(random_array)
poker_test_with_chi_square_test(random_array)
consecutive_runs_test(random_array)
````
